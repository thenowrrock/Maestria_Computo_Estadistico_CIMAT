### Problema de suicidios

```python
suicide_data = pd.read_csv("suicide_data.csv")
suicide_data.head(5)
```

```python
suicide_data.columns
```

```python
suicide_data = suicide_data.rename(columns={'suicides/100k pop':'suicides_100k_pop', 'HDI for year':'hdi_for_year',
       ' gdp_for_year ($) ':'gdp_for_year', 'gdp_per_capita ($)':'gdp_per_capita'})
```

```python
suicide_data.info()
```

Observamos que un **gran porcentaje** de los registros no tienen el HDI.

```python
hdi_null = suicide_data.groupby("year",as_index=False).hdi_for_year.count()
```

```python
fig = plt.figure(figsize = (4,2))
ax=fig.add_axes([0,0,1,1])
plt.bar(hdi_null["year"], hdi_null["hdi_for_year"])
#plt.title("Total de registros de HDI por ao")
plt.xlabel("ao")
plt.ylabel("nmero de registros")
plt.savefig('figs/dhi_null.png')
```

```python
sns.set()
fig = plt.figure(figsize = (4,2))
ax=fig.add_axes([0,0,1,1])
count_country = suicide_data.groupby(["year","country"], as_index=False).sex.count().groupby("country", as_index=False).year\
    .count().sort_values("year", ascending=False)
count_country.year.hist()
plt.xlabel("nmero de registros")
plt.ylabel("nmero de pases")
#plt.title("Nmero de registro de los pases")
plt.show()
```

### Tendencia de los suicidios por pases

```python
# suicide_data = suicide_data[suicide_data.country.isin(list_country)]
```

```python
suicide_country = suicide_data.groupby(["country", "year"], as_index=False).agg({'suicides_no': 'sum', 
                                                                   'population': 'sum',
                                                                    'gdp_per_capita':'mean',
                                                                    'hdi_for_year':'mean'})

suicide_country["suicides/100k_pop"] = suicide_country["suicides_no"]*100000/suicide_country["population"]
suicide_country["year_dt"] = pd.to_datetime(suicide_country["year"].astype(str))
```

```python
suicide_country.gdp_per_capita.hist()
```

```python
suicide_country["class_gdp"] = suicide_country.gdp_per_capita.apply(
    lambda x: "low_income" if x<1026 else ("low_middle_income" if x<3995 else 
                                         ("upper_middle_income" if x<12375 else "high_income")))
```

```python
suicide_country.groupby("class_gdp", as_index=False).country.count()
```

```python
fig = plt.figure(figsize = (10,6))
ax=fig.add_axes([0,0,1,1])
sns.stripplot(x='year', y='suicides/100k_pop', hue="class_gdp", data=suicide_country,   
              palette='viridis', jitter=True, edgecolor='none')
plt.xticks(rotation=90)
plt.ylabel("tasa de suicidios (100 mil hab)")
plt.xlabel("ao")
plt.show()
```

```python
country_whiout_hdi= suicide_country[~suicide_country.hdi_for_year.isnull()]

import matplotlib.colors as mcolors
import matplotlib.cm as cm

fig = plt.figure(figsize = (10,6))
ax=fig.add_axes([0,0,1,1])
g = sns.stripplot(x='year', y='suicides/100k_pop', hue="hdi_for_year", data=country_whiout_hdi,   
              palette='viridis', jitter=True, edgecolor='none')
g.legend([],[], frameon=False)
sns.despine()
# Drawing the side color bar
normalize = mcolors.Normalize(vmin=country_whiout_hdi['hdi_for_year'].min(), 
                              vmax=country_whiout_hdi['hdi_for_year'].max())
colormap = cm.viridis

for n in country_whiout_hdi['hdi_for_year']:
    plt.plot(color=colormap(normalize(n)))

scalarmappaple = cm.ScalarMappable(norm=normalize, cmap=colormap)
scalarmappaple.set_array(country_whiout_hdi['hdi_for_year'])
plt.colorbar(scalarmappaple)
plt.ylabel("tasa de suicidios (100 mil hab)")
plt.xlabel("ao")
plt.show()
```
